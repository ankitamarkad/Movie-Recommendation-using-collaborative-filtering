```python
#convert into a sparse matrix
from scipy.sparse import csr_matrix
```

```python
#create pivot table
movie_user = df_ratings.pivot(index = 'movieId',columns='userId',values = 'rating')
```

```python
movie_user
```

```python
#since most of the users have not rated all the movies, the respective fields are null/NaN
```

```python
movie_user = df_ratings.pivot(index = 'movieId',columns='userId',values = 'rating').fillna(0)
```

```python
movie_user
```

```python
matrix_movies_users = csr_matrix(movie_user.values)
```
